```python
#generate a random signal and then alpha dyscynronlize it then plot both and plot the power spectral density 
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch
# Generate a random signal
np.random.seed(0)
fs = 1000  # Sampling frequency
t = np.arange(0, 1.0, 1.0/fs)
signal = np.random.normal(0, 1, len(t))
# Alpha dyscynronlization (simulated by applying a band-stop filter around 8-12 Hz)
from scipy.signal import butter, filtfilt
def bandstop_filter(data, lowcut, highcut, fs, order=4):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='bandstop')
    y = filtfilt(b, a, data)
    return y
dyscyn_signal = bandstop_filter(signal, 8, 12, fs)
# Plot both signals
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, signal, label='Original Signal')
plt.plot(t, dyscyn_signal, label='Alpha Dyscynronlized Signal', alpha=0.7)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Original and Alpha Dyscynronlized Signals')
plt.legend()
# Plot Power Spectral Density
plt.subplot(2, 1, 2)
f_orig, Pxx_orig = welch(signal, fs, nperseg=256)
f_dyscyn, Pxx_dyscyn = welch(dyscyn_signal, fs, nperseg=256)
plt.semilogy(f_orig, Pxx_orig, label='Original Signal PSD')
plt.semilogy(f_dyscyn, Pxx_d    yscyn, label='Dyscynronlized Signal PSD', alpha=0.7)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power/Frequency [V**2/Hz]')
plt.title('Power Spectral Density')
plt.legend()
plt.tight_layout()
plt.show()yscyn, label='Dyscynronlized Signal PSD', alpha=0.7)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power/Frequency [V**2/Hz]')
plt.title('Power Spectral Density')
plt.legend()
plt.tight_layout()
plt.show()  
```
